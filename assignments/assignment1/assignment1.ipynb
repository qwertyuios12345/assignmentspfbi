```python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]
```

```python
#part a
```

```python
valid_transactions = []
for transaction in transactions: 
    if transaction["price"] > 0 and transaction["quantity"] >= 0: 
        valid_transactions.append(transaction)
for i in valid_transactions:
    print(i)
```

```python
#part b
```

```python
for total in valid_transactions:
    total["total"] = total["price"] * total["quantity"]
print(valid_transactions)
```

```python
#part c
```

```python
revenue_per_region = {}
for region in valid_transactions:
    if region["region"] in revenue_per_region: 
        revenue_per_region[region["region"]] += region["total"]
    else:
        revenue_per_region[region["region"]] = region["total"]
print(revenue_per_region)
```

```python
#part d
```

```python
top_region = max(revenue_per_region, key=Revenue_per_region.get)
print(f"Highest total revenue: {top_region} = {Revenue_per_region[top_region]}")
```

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

#function

def summarise_sales(saleslist):
    total = sum(saleslist)
    average = round(total/len(saleslist),2)
    best_day = saleslist.index(max(saleslist)) + 1
    worst_day = saleslist.index(min(saleslist)) + 1
    above_average_count = 0
    for day in saleslist: 
        if day > average:
            above_average_count += 1

#dictionary

    return{
        "total": total,
        "average": average,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": above_average_count
}

print(summarise_sales(sales))
```

```python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}
#task a
    def add_asset(self, ticker, shares, price_per_share):
        if ticker in self.holdings:
            self.holdings[ticker]["shares"] += shares
            self.holdings[ticker]["price_per_share"] = price_per_share
        else:
            self.holdings[ticker] = {"shares": shares, "price_per_share": price_per_share}

#task b
    def total_value(self):
        total = 0
        for ticker in self.holdings:
            total += self.holdings[ticker]["shares"] * self.holdings[ticker]["price_per_share"]
        return total

#task c
    def most_valuable(self):
        return max(self.holdings, key = lambda ticker : self.holdings[ticker]["shares"] * self.holdings[ticker]["price_per_share"])

#task d
    def __str__(self):
        return f" Owner: {self.owner}, assets = {len(self.holdings)}, total value = {self.total_value()}"
            

my_portfolio = Portfolio("John") 

my_portfolio.add_asset("AAPL", 10, 150)
my_portfolio.add_asset("TSLA", 5, 200)
my_portfolio.add_asset("NVDA", 8, 300)

print(my_portfolio.total_value())
print(my_portfolio.most_valuable())
print(my_portfolio)
```

```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)

#part a

df = pd.DataFrame({
    "date": days,
    "visitors": visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value
})

df = df.set_index("date")

#part b

df["conversions"] = round(df["visitors"] * df["conversion_rate"])

df["daily_revenue"] = df["conversions"] * df["avg_order_value"]


print(df)
```
